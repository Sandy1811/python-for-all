```python
# Defining the X and Y variables:

X = salary[["Years of Experience"]]

y = salary["Salary"]
```

```python
# Splitting the dataset into Training and testing set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.25)
```

```python
X_test.shape
```

```python
# Implementing the ML Model

from sklearn.linear_model import LinearRegression

regressor = LinearRegression(fit_intercept=True)

regressor.fit(X_train, y_train)
```

```python
print("Linear Model Coeff %s" % regressor.coef_)

print("linear Model Intercept %s" % regressor.intercept_)
```

```python
X_test_prediction =  regressor.predict(X_test)
```

```python
csv = pd.DataFrame(X_test_prediction)

csv.to_csv("X_test_prediction.csv")
```

```python
X_test.to_csv("X_test.csv")
```

```python
plt.scatter(X_train, y_train, color = "green")
plt.plot(X_train, regressor.predict(X_train), color = "red")
plt.ylabel('Salary/Year [dollars]')
plt.xlabel('Years of Experience')
plt.title('Salary vs. Years of Experience (Training dataset)')
```
